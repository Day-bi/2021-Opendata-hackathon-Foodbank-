```python
# 선호물품수량이 가장 많은 상위 30개 센터
##센터코드+선호코드+수량 -> raw
df = test.drop(columns=['Unnamed: 0','areaCd','unitySignguCd'], axis=1)
df = df[df['holdQy'] != 0]
df.sort_values(by = "holdQy",
              inplace = True,
              ascending = False)
Qy_30_raw = df.head(31)
Qy_30_raw.to_csv("Qy_30_raw.csv") 
```

```python
# 선호물품수량이 가장 많은 상위 30개 센터
#센터코드+선호코드+수량 -> label
df = test.drop(columns=['Unnamed: 0','areaCd','unitySignguCd'], axis=1)
df.columns = ['센터코드','선호물품분류코드','보유수량']
df = df[df['보유수량'] != 0]
df.sort_values(by = "보유수량",
              inplace = True,
              ascending = False)

# 센터코드 데이터
spctrCd = pd.read_csv('spctrCd.csv')
preferCnttgClscd = pd.read_csv('preferCnttgClscd.csv')
preferCnttgClscd.columns = ['선호물품분류코드','선호물품분류명']

# 데이터 병합
m_df = pd.merge(df,spctrCd)
m_df = pd.merge(m_df,preferCnttgClscd)
m_df = m_df.drop(columns = ['센터코드','선호물품분류코드'], axis=1)
m_df = m_df[['센터코드명','선호물품분류명','보유수량']]
m_df.sort_values(by = "보유수량",
              inplace = True,
              ascending = False)

# 상위 30개
Qy_30_label = m_df.head(31)
Qy_30_label.to_csv("Qy_30_label.csv")
```
